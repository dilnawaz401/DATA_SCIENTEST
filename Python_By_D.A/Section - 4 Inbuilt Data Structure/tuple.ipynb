```python
## immutable nature of tuple
## elements of tuple cannot be changed, added or removed
lst = [1, 2, 3, 4, 5]
print (lst)

lst[0] = 10
print (lst)
```

```python
 ## immutable nature 
 
number = (1, 2, 3, 4, 5)
print (number)
number[0] = 10
```

```python
## tuple method
print (number.count(1)) # 1
print (number.index(3)) # 2
```

```python
## packing and unpacking of tuple
## packing
packed_tuple = 1, 2, 3, 4, 5
print (packed_tuple)

## unpacking

a, b, c, d, e = packed_tuple
print (a) # 1
print (b) # 2
print (c) # 3
```

```python
## unpacking with * 
number = (1, 2, 3, 4, 5)
first, *middle, last = number
print (first) # 1
print (middle) # [2, 3, 4]
print (last) # 5
```

```python
##nested tuple
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print (lst[0]) # [1, 2, 3]
print (lst[0][1]) # 2
print (lst[0][0:3]) # 2
```

```python
## nested tuple
nested_tuple = ((1, 2) , (3, 4), (5, 6),(3,True, False))
print (nested_tuple[0]) # (1, 2
print (nested_tuple[3][1]) # 2
print (nested_tuple[3][0:3]) # 2
```
